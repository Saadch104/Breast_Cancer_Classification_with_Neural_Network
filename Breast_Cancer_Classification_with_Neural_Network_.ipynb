```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn.datasets
from sklearn.model_selection import train_test_split
```

```python
#loading the breast cancer dataset from sklearn datasets
breast_cancer_dataset = sklearn.datasets.load_breast_cancer()
```

```python
print(breast_cancer_dataset)   #contain feature values ....0 and 1 are the target (Malignant )and(Benign)
```

```python
#loading the datatset into dataframe
data_frame= pd.DataFrame(breast_cancer_dataset.data , columns=  breast_cancer_dataset.feature_names)
```

```python
#adiing the "Target " column to the dataframe
data_frame['label']=breast_cancer_dataset.target
```

Sperating the targets and Features

```python
X= data_frame.drop(columns = 'label' , axis= 1)
#when you drop a row you will write axis = 0
Y= data_frame['label']
```

```python
print(X)
```

```python
print(Y)
```

Splitting data into train and test

```python
x_train,x_test,y_train,y_test= train_test_split(X,Y,test_size=0.2,random_state=2)
```

```python
print(X.shape , x_train.shape ,x_test.shape)
```

```python
#preprocessing // standard scalar
from sklearn.preprocessing import StandardScaler

scaler= StandardScaler()
x_train_std=scaler.fit_transform(x_train)
x_test_std = scaler.transform(x_test)
```

```python
#importing libraries
import tensorflow as tf
tf.random.set_seed(3)#its a parameter...no matter how many time you run the code you will get the same accuracy score.
from tensorflow import keras
```

```python
#building a NN

model = keras.Sequential([
    keras.layers.Flatten(input_shape=(30,)),#(convrt the data into single-D array)no of neurons in input is equal to no of features in dataset(31)
    keras.layers.Dense(20, activation='relu'),
    keras.layers.Dense(2 , activation='sigmoid')
])
```

```python
#compiling
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
history= model.fit(x_train_std,y_train, epochs=10, validation_split=0.1)
```

```python
#visualization
#for accuracy
plt.plot(history.history['accuracy']  )
plt.plot(history.history['val_accuracy']  )

plt.title("Model Accuracy")
plt.xlabel("Epochs")
plt.ylabel("ACcuracy")

plt.legend(['Train,val'], loc = 'lower right')



```

```python
#for loss

plt.plot(history.history['loss']  )
plt.plot(history.history['val_loss']  )

plt.title("Model loss")
plt.xlabel("Epochs")
plt.ylabel("loss")

plt.legend(['Train data , validation'])
```

# **Accuracy of model on test data**

```python
loss , accuracy = model.evaluate(x_test_std, y_test)#evaluate function itself teake the X_test_std data and compare it with the y_test data .Predict the labels and return 2 values loss and accuracy
print("Accuracy is = ",accuracy)
print("loss is = ",loss)
```
